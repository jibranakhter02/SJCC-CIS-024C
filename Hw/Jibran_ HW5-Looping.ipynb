#### Exercise 1 - Parsing a comma separated set of values

Use a **while** loop to ask the user to enter a list of items for a grocery list

Ensure that each item in the grocery list is separated by a comma. Use the **split** command to extract each token (item) in the grocery list. Print the last item in the list.

Ask the user if the user wants to continue. If no, terminate the loop and display the message 'Program ended'

**Example**

```
Enter the grocery list: cookies, cream, sugar
Last item: sugar

Do you want to continue (y/n): y

Enter a line of text: veggies, yogurt
Last item: yogurt

Do you want to continue (y/n): n
Program ended


````python
### YOUR CODE GOES
while True:
    grocery_list = input("Enter the grocery list (separate items by commas): ")

    items = grocery_list.split(",")

    print(f"Last item: {items[-1].strip()}")

    continue_choice = input("Do you want to continue (y/n): ").lower()

    if continue_choice == 'n':
        print("Program ended")
        break
### END CODE
````

#### Exercise 2

Ask the user to type in a grocery list and a search item. Ensure that each item in the grocery list is separated by a comma and an arbitrary number of spaces. Use the **split** command to search for the search item in this list.

For example, let us say that the user enters ""milk ,  bananas,  sugar,  eggs,  cheese  " (notice the arbitrary spaces between items) and the search term is "eggs". You will need to look for "eggs" in the grocery list and if found, print the message "Item found", otherwise, print "Item not found"

````python
### YOUR CODE GOES
grocery_list = input("Enter the grocery list (separate items by commas): ")

search_item = input("Enter the item to search: ").strip()

items = [item.strip() for item in grocery_list.split(",")]

if search_item in items:
    print("Item found")
else:
    print("Item not found")
### END CODE
````

#### Exercise 3

Write a python program that takes in a list of words from the user and prints the shortest word in the list. If two words are equal short, just pick the first one that you see.


````python
### YOUR CODE GOES
word_list = input("Enter a list of words (separated by spaces): ")

words = word_list.split()

shortest_word = min(words, key=len)

print(f"The shortest word is: {shortest_word}")
### END CODE
````

#### Exercise *4* - Displaying values in a list object

Create a list object with 10 arbitrary numbers. Use a **for loop** to display only the odd numbers in the list.

**Example:**

If your list has ```[2,3,4,5,6]```, then you will need to display ```[3,5]```


````python
### YOUR CODE GOES
numbers = [12, 5, 8, 13, 21, 9, 2, 7, 14, 11]

odd_numbers = []
for num in numbers:
    if num % 2 != 0:
        odd_numbers.append(num)

print("Odd numbers:", odd_numbers)
### END CODE
````

#### Exercise *5* - Printing without having to repeat the logic of printing

Write a program that prints the even numbers 2 to 21 but without having to repeat your logic of printing 10 different times. You can use a while loop.

````python
### YOUR CODE GOES
number = 2

while number <= 21:
    print(number)
    number += 2
### END CODE
````

#### Exercise *6* - Printing even and odd numbers between 1 to 50

Print odd and even numbers between 1 to 50.
Range starts at 1, ends at stopping point 50
step is 4 which means each time range value will be incremented by 4

````python
### YOUR CODE GOES
print("Odd and even numbers between 1 and 50 with a step of 4:")

for num in range(1, 51, 4):
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
### END CODE
````
