**Example:** Sort the list alphabetically:

#### Exercise 5 - Updating the values of a tuple

colors = ('blue', 'red', 'green', 'white', 'purple')

You want to:

1 - change the 'white' color to 'gray'

2 - add 'yellow' color to the tuple

3 - print the sorted tuple alphabetically

````python
### YOUR CODE GOES BELOW
colors = ('blue', 'red', 'green', 'white', 'purple')

colors_list = list(colors)

colors_list[colors_list.index('white')] = 'gray'

colors_list.append('yellow')

colors = tuple(colors_list)

print(tuple(sorted(colors)))
### END CODE
````
