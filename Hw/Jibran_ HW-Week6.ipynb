#### Exercise 1 - Max of three numbers
*  Write a Python function to find the maximum of three numbers in a list.
*  Write a Python function to find the minimum of three numbers in a list.
*  Write a Python function to return the last number in a list.
*  Write a Python function to return the first number in a list of numbers

The function must accept one list parameter. You cannot pass in 3 separate numbers to the function

Make sure you invoke this function in your main program and display the result

*Note that you must write your own function and not use any of the prebuilt functions like max or min *

````python
### YOUR CODE GOES
def max_of_three(numbers):
    max_num = numbers[0]
    for num in numbers[1:]:
        if num > max_num:
            max_num = num
    return max_num

def min_of_three(numbers):
    min_num = numbers[0]
    for num in numbers[1:]:
        if num < min_num:
            min_num = num
    return min_num

def last_number(numbers):
    return numbers[-1]

def first_number(numbers):
    return numbers[0]

if __name__ == "__main__":
    numbers = [25, 14, 32]

    max_result = max_of_three(numbers)
    min_result = min_of_three(numbers)
    last_result = last_number(numbers)
    first_result = first_number(numbers)

    print(f"Maximum of three numbers: {max_result}")
    print(f"Minimum of three numbers: {min_result}")
    print(f"Last number in the list: {last_result}")
    print(f"First number in the list: {first_result}")
### END CODE
````

#### Exercise 2 - Finding the sum of numbers in a list

Write a Python function to sum all the numbers in a list

Make sure you invoke this function in your main program and display the result

Use a while loop to continue asking the user till the user decides to end.

**Example**
```
Enter a list of numbers: 1,2,3
The product is 6

Do you want to continue(y/n): y

Enter a list of numbers: 2,3
The product is 5

Do you want to continue(y/n): n
Program ended
```

````python
### YOUR CODE GOES
def sum_of_numbers(numbers):
    total = 0
    for num in numbers:
        total += num
    return total

def main():
    while True:
        numbers_input = input("Enter a list of numbers (separated by commas): ")
        numbers = [int(x) for x in numbers_input.split(',')]

        sum_result = sum_of_numbers(numbers)

        print(f"The sum is {sum_result}")

        continue_choice = input("Do you want to continue (y/n): ").lower()
        if continue_choice != 'y':
            print("Program ended")
            break

main()
### END CODE
````

#### Exercise 3 - Finding the product of numbers in a list

Ask the user to enter a list of numbers. Write a Python function to multiply all the numbers in a list

Make sure you invoke this function in your main program and display the result

Use a while loop to continue asking the user till the user decides to end.

**Example**
```
Enter a list of numbers: 1,2,3
The product is 6

Do you want to continue(y/n): y

Enter a list of numbers: 2,3
The product is 6

Do you want to continue(y/n): n
Program ended
```

````python
### YOUR CODE GOES BELOW
def product_of_numbers(numbers):
    product = 1
    for num in numbers:
        product *= num
    return product

def main():
    while True:
        numbers_input = input("Enter a list of numbers (separated by commas): ")
        numbers = [int(x) for x in numbers_input.split(',')]

        product_result = product_of_numbers(numbers)

        print(f"The product is {product_result}")

        continue_choice = input("Do you want to continue (y/n): ").lower()
        if continue_choice != 'y':
            print("Program ended")
            break

main()
### END CODE
````

Exercise 4 - Finding the factorial of a number without recursion
Write a Python function to calculate the factorial of a number (a non-negative integer). The function accepts the number as an argument.

Use a while loop to continue asking the user till the user decides to end.

Example

Enter a number: 6
Factorial of 6 is 720

Do you want to continue(y/n): y

Enter a number: 4
Factorial of 4 is 24

Do you want to continue(y/n): n
Program ended

````python
### YOUR CODE GOES BELOW
def factorial(number):
    result = 1
    for i in range(1, number + 1):
        result *= i
    return result

def main():
    while True:
        number_input = int(input("Enter a number: "))

        factorial_result = factorial(number_input)

        print(f"Factorial of {number_input} is {factorial_result}")

        continue_choice = input("Do you want to continue (y/n): ").lower()
        if continue_choice != 'y':
            print("Program ended")
            break

main()
### END CODE
````

#### Exercise 5 - Check if number falls in a given range

Write a Python function to check whether a number falls in a given range. The function accepts 3 parameters, the lower limit of the range, the upper limit of the range, and the number.

Use a while loop to continue asking the user till the user decides to end.

The function parameters should be as follows
```
def checkNumberInRange(lowerLimit, upperLimit, number):
```


**Example**
```
Enter the lower limit: 2
Enter the upper limit: 200
Enter the number: 100
The number falls within the range

Do you want to continue(y/n): y

Enter the lower limit: 2
Enter the upper limit: 200
Enter the number: 1000
The number falls outside the range

Do you want to continue(y/n): n
Program ended
```



````python
### YOUR CODE GOES BELOW
def checkNumberInRange(lowerLimit, upperLimit, number):
    if lowerLimit <= number <= upperLimit:
        return "The number falls within the range"
    else:
        return "The number falls outside the range"

def main():
    while True:
        lower_limit = int(input("Enter the lower limit: "))
        upper_limit = int(input("Enter the upper limit: "))
        number = int(input("Enter the number: "))

        result = checkNumberInRange(lower_limit, upper_limit, number)

        print(result)

        continue_choice = input("Do you want to continue (y/n): ").lower()
        if continue_choice != 'y':
            print("Program ended")
            break

main()
### END CODE
````
