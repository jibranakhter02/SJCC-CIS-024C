Python program to find maximum between three numbers:

```python
def find_maximum(num1, num2, num3):
    if num1 >= num2 and num1 >= num3:
        return num1
    elif num2 >= num1 and num2 >= num3:
        return num2
    else:
        return num3

num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
num3 = int(input("Enter the third number: "))

maximum = find_maximum(num1, num2, num3)
print(f"The maximum number is: {maximum}")
```

Python program to check whether a number is negative, positive or zero:

```python
def check_number(num):
    if num > 0:
        return "Positive"
    elif num < 0:
        return "Negative"
    else:
        return "Zero"

number = int(input("Enter a number: "))

result = check_number(number)
print(f"The number is: {result}")
```

Python program to check whether a number is divisible by 5 and 11 or not:

```python
def check_divisibility(num):
    if num % 5 == 0 and num % 11 == 0:
        return "The number is divisible by both 5 and 11."
    else:
        return "The number is not divisible by both 5 and 11."

number = int(input("Enter a number: "))

result = check_divisibility(number)
print(result)
```

Python program to check whether a number is even or odd:

```python
def check_even_odd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

number = int(input("Enter a number: "))

result = check_even_odd(number)
print(f"The number is: {result}")
```
