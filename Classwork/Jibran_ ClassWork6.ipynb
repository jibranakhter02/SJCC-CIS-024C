### **ClassWork6.2:**

Write an function, register(), that asks the user to accept to register for python class this term, reads in Y/N, and outputs a response by calling accepted() or rejected(). accepted() prints "Thank you for taking python class this term." and rejected() prints "You have rejected to take python this term. Thank you.".

Given inputs 2, "Y" and "N", the output is:

Do you accept to register for python class this term?
Y
Thank you for registering for python, this term.
Given a function call to register() and input "N", the output is:

Do you accept to register for python class this term?
N
You have rejected to take python this term. Thank you.

```python
def accepted():
    print("Thank you for registering for python, this term.")

def rejected():
    print("You have rejected to take python this term. Thank you.")

def register():
    response = input("Do you accept to register for python class this term? (Y/N) ")

    if response.upper() == 'Y':
        accepted()
    elif response.upper() == 'N':
        rejected()
    else:
        print("Invalid input. Please enter Y or N.")
register()
```
