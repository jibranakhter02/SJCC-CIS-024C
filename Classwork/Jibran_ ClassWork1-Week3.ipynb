# **Python String format() Method**

Example:
Insert the price inside the placeholder, the price should be in fixed point, two-decimal format:

txt = "For only {price:.2f} dollars!"

print(txt.format(price = 49))



```python
txt = "For only {price:.2f} dollars!"
print(txt.format(price = 49))
```

# **The Placeholders**
The placeholders can be identified using:

*   named indexes {price},
*   numbered indexes {0},
*   empty placeholders {}.

Example:
Using different placeholder values:

txt1 = "My name is {fname}, I'm {age}".format(fname = "John", age = 36)

txt2 = "My name is {0}, I'm {1}".format("John",36)

txt3 = "My name is {}, I'm {}".format("John",36)

print(txt1)

print(txt2)

print(txt3)

```python
txt1 = "My name is {fname}, I'm {age}".format(fname = "John", age = 36)
txt2 = "My name is {0}, I'm {1}".format("John",36)
txt3 = "My name is {}, I'm {}".format("John",36)

print(txt1)
print(txt2)
print(txt3)
```

# **Formatting Types**
Inside the placeholders you can add a formatting type to format the result:

Cl;assWork1-Week3.png

Ref. https://www.w3schools.com/python/ref_string_format.asp

```python
txt = "We have {:<7} edibles."

print(txt.format(50))
```

```python
txt = "We have {:>9} computers."

print(txt.format(100))
```

```python
txt = "We have {:^6} drinks."

print(txt.format(10))
```

```python
txt = "The temperature is {:=20} degrees celsius."

print(txt.format(-10))
```

```python
txt = "The temperature is between {:+} and {:+} degrees celsius."

print(txt.format(-2, 14))
```

```python
txt = "The temperature is between {:-} and {:-} degrees celsius."

print(txt.format(-2, 8))
```

```python
txt = "The temperature is between {: } and {: } degrees celsius."

print(txt.format(-3, 7))
```

```python
txt = "The universe is {:,} years old."

print(txt.format(13850506000))
```

```python
txt = "The universe is {:_} years old."

print(txt.format(13850506000))
```

```python
txt = "The binary version of {0} is {0:b}"

print(txt.format(10))
```

```python
txt = "The unicode version of {0} is {0:c}"

print(txt.format(999))
```

```python
txt = "We have {:d} edibles."

print(txt.format(0b101))
```

```python
txt = "We have {:e} chickens."

print(txt.format(5.674563728))
```

```python
txt = "We have {:E} chickens."

print(txt.format(5.54673829))
```

```python
txt = "The price is {:.2f} dirhams."

print(txt.format(24))

txt = "The price is {:f} dirhams."

print(txt.format(24))
```

```python
x = float('inf')

txt = "The price is {:F} rupees."

print(txt.format(x))

txt = "The price is {:f} rupees."

print(txt.format(x))
```

```python
x = float('inf')

txt = "The price is {:g} dollars."

print(txt.format(x))
```

```python
x = float('inf')

txt = "The price is {:G} dollars."

print(txt.format(x))
```

```python
txt = "The octal version of {0} is {0:o}"

print(txt.format(15))
```

```python
txt = "The Hexadecimal version of {0} is {0:x}"

print(txt.format(260))
```

```python
txt = "The Hexadecimal version of {0} is {0:X}"

print(txt.format(230))
```

```python
txt = "The number format of {0} is {0:n}"

print(txt.format(4.567))
```

```python
txt = "You scored {:%}"

print(txt.format(0.70))

txt = "You scored {:.0%}"

print(txt.format(0.70))
```
