  1-Calculate the product of elements in the list, number, using a for loop. Then print the product.

number = [2, 3, 4, 6, 8, 9, 14, 22, 54 ]

```python
#your code
product=1
number = [2, 3, 4, 6, 8, 9, 14, 22, 54 ]
for num in number:
  product *= num

print(product)
#End  code
```

  2- Find the largest number in a list using a for loop:

number = [3, 9, 1, 6, 2, 8, 5]

```python
#your code
number = [3, 9, 1, 6, 2, 8, 5]
largest = number[0]

for num in number:
    if num > largest:
        largest = num

print("The largest number is:", largest)
#End  code
```

   3- Find the average of numbers in a list using a for loop:

number = [2, 3, 4, 6, 8, 9, 14, 22, 54 ]

```python
#your code
number = [2, 3, 4, 6, 8, 9, 14, 22, 54]
total = 0

for num in number:
    total += num

average = total / len(number)
print("The average is:", average)
#End  code
```

   4- Find the first occurrence of a number(target) in a list using a while loop:


   
   number = [3, 8, 2, 7, 4, 9]
   
   target = 7

```python
#your code
number = [3, 8, 2, 7, 4, 9]
target = 7
index = 0

while index < len(number):
    if number[index] == target:
        print("Target found at index:", index)
        break
    index += 1
else:
    print("Target not found.")
#End  code
```
