### **Basic Recursion Questions:**

**I** - Counting Down

A simple recursive function that prints numbers from n down to 1

```python
def countdown(n):
  if n < 1:
    return

  print(n)

  countdown(n - 1)

print("Counting down from 5:")
countdown(5)
```

**II** - Sum of numbers from 1 to n
This function calculates the sum of numbers from 1 to n using recursion

```python
def sum_num(n):
  if n <= 0:
    return 0
  return n + sum_num(n - 1)

num = int(input("Enter a number: "))
result = sum_num(num)
print("The sum of numbers from 1 to", num, "is: ", result)
```

**III** - Calculation the power of a number

A ricursive function to calculate a^b.

```python
def power(a, b):
  if b == 0:
    return 1
  elif b > 0:
    return a * power(a, b-1)

base = float(input("Enter base number: "))
exponent = int(input("Enter the exponent: "))

result = power(base, exponent)
print(result)
```

**IV**- A recursive functionto find the length of a list.

```python
def recursive_length(lst):
  if lst == []:
    return 0
  else:
    return 1 + recursive_length(lst[1:])

example_list = [1, 2, 3, 4, 5]
print("Length of the list:", recursive_length(example_list))
```

**V** - Checking if a Srting is Palindrome

A recursive function to check if a String reads the same forwars and backward.

```python
def palindrome(s, i=0):

  if i > len(s)/2:
    return True

  if s[i] == s[len(s) - i - 1] and palindrome(s, i + 1):
    return True
  return False

print("Let's see if your word is a palindrome:" )
message = input("Enter your word: ").lower()
if palindrome(message):
  print(f"Yes, it is a Palindrome {message}")
else:
  print(f"No, it is not a Palindrome {message}")

```
