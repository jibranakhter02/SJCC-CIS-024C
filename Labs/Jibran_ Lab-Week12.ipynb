1) Write a Python program to sum recursion lists using recursion.

Test Data: [1, 2, [3,4], [5,6]]

Expected Result: 21

```python
def recursive_sum(data):
    total = 0
    for element in data:
        if isinstance(element, list):
            total += recursive_sum(element)
        else:
            total += element
    return total

data = [1, 2, [3, 4], [5, 6]]

result = recursive_sum(data)
print(f"The sum is: {result}")
```

2) Write a Python program to get the sum of a non-negative integer using recursion.

Test Data:

sumDigits(345) -> 12

sumDigits(45) -> 9

```python
def sumDigits(n):
    if n == 0:
        return 0
    else:
        return n % 10 + sumDigits(n // 10)

# Test Data
print(f"sumDigits(345) -> {sumDigits(345)}")
print(f"sumDigits(45) -> {sumDigits(45)}")
```

3) Write a Python program to calculate the sum of the positive integers of n+(n-2)+(n-4)... (until n-x =< 0) using recursion .

Test Data:

sum_series(6) -> 12

sum_series(10) -> 30

```python
def sum_series(n):
    if n <= 0:
        return 0
    else:
        return n + sum_series(n - 2)

print(f"sum_series(6) -> {sum_series(6)}")
print(f"sum_series(10) -> {sum_series(10)}")
```

4)  Write a Python program to find the greatest common divisor (GCD) of two integers using recursion.

```python
def gcd(a, b):
    if b == 0:
        return a
    else:
        return gcd(b, a % b)

# Test Data
print(f"gcd(48, 18) -> {gcd(48, 18)}")
print(f"gcd(56, 98) -> {gcd(56, 98)}")
```
