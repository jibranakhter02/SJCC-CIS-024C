### Exercise 1
Write a Python program that simulates a pair of dice for the user

**Algorithm**

Loop till user wants to stop the dice roll.

Simulate two dice roll.

Display the two numbers

Reference: https://docs.python.org/3/library/random.html

```python
### YOUR CODE GOES BELOW
import random

def roll_dice():
    """Simulate rolling two dice."""
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    return die1, die2

def main():
    while True:
        die1, die2 = roll_dice()
        print(f"You rolled: {die1} and {die2}")

        choice = input("Do you want to roll again? (yes/no): ").lower()
        if choice != 'yes':
            break

main()
### END CODE
```

#### Exercise 2

Write a Python program to convert the temperature for the below

* Fahrenheit to Centigrade and vice versa
* Fahrenheit to Kelvin and vice versa
* Kelvin to Centigrade and vice versa

You should write functions to do each of these operations. The functions should be present in an external .py file named helper.py

Your main program will do the following

**Algorithm**

* Import your module
* Loop till use is done
    * Ask user to select an option from the 6 available options
    * Get the input temperature value
    * Display the converted temperature value



```python
### YOUR CODE GOES BELOW
def fahrenheit_to_centigrade(f):
    return (f - 32) * 5 / 9

def centigrade_to_fahrenheit(c):
    return (c * 9 / 5) + 32

def fahrenheit_to_kelvin(f):
    return (f - 32) * 5 / 9 + 273.15

def kelvin_to_fahrenheit(k):
    return (k - 273.15) * 9 / 5 + 32

def kelvin_to_centigrade(k):
    return k - 273.15

def centigrade_to_kelvin(c):
    return c + 273.15


def main():
    while True:
        print("\nTemperature Conversion Options:")
        print("1: Fahrenheit to Centigrade")
        print("2: Centigrade to Fahrenheit")
        print("3: Fahrenheit to Kelvin")
        print("4: Kelvin to Fahrenheit")
        print("5: Kelvin to Centigrade")
        print("6: Centigrade to Kelvin")

        option = int(input("Select an option (1-6): "))
        temperature = float(input("Enter the temperature: "))

        if option == 1:
            print(f"Converted: {fahrenheit_to_centigrade(temperature)} °C")
        elif option == 2:
            print(f"Converted: {centigrade_to_fahrenheit(temperature)} °F")
        elif option == 3:
            print(f"Converted: {fahrenheit_to_kelvin(temperature)} K")
        elif option == 4:
            print(f"Converted: {kelvin_to_fahrenheit(temperature)} °F")
        elif option == 5:
            print(f"Converted: {kelvin_to_centigrade(temperature)} °C")
        elif option == 6:
            print(f"Converted: {centigrade_to_kelvin(temperature)} K")

        cont = input("Do you want to convert another temperature? (yes/no): ").lower()
        if cont != 'yes':
            break

main()
### END CODE
```

### Exercise 3
Write a Python program to find the areas of the shapes of the following figures

* Circle
* Square
* Rectangle

You should write functions to do each of these operations. The functions should be present in an external .py file named helper.py

Your main program will do the following

**Algorithm**

* Import your module
* Loop till use is done
* Ask user to select an option from the 3 available options
* Get the input value from the user depending on the shape
* Display the area of the shape using the functions in your helper module

```python
### YOUR CODE GOES BELOW
import math

def area_circle(radius):
    return math.pi * radius ** 2

def area_square(side):
    return side ** 2

def area_rectangle(length, width):
    return length * width


def main():
    while True:
        print("\nArea Calculation Options:")
        print("1: Circle")
        print("2: Square")
        print("3: Rectangle")

        option = int(input("Select a shape (1-3): "))

        if option == 1:
            radius = float(input("Enter the radius of the circle: "))
            print(f"Area of the circle: {area_circle(radius)}")
        elif option == 2:
            side = float(input("Enter the side of the square: "))
            print(f"Area of the square: {area_square(side)}")
        elif option == 3:
            length = float(input("Enter the length of the rectangle: "))
            width = float(input("Enter the width of the rectangle: "))
            print(f"Area of the rectangle: {area_rectangle(length, width)}")

        cont = input("Do you want to calculate another area? (yes/no): ").lower()
        if cont != 'yes':
            break

main()
### END CODE
```
