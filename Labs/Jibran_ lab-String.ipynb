### Exercise 5 (4 points)
Write a Python program to get a list of your fruits (more than 5 fruits). Then find the legth of your list. Modify your list by removing the 3rd item ffrom the list.

```python
### YOUR CODE GOES BELOW
fruits = input("Enter a list of fruits (more than 5), separated by commas: ")

fruit_list = fruits.split(",")

list_length = len(fruit_list)
print(f"The length of your list is: {list_length}")

if list_length >= 3:
    removed_fruit = fruit_list.pop(2)
    print(f"Removed the 3rd fruit: {removed_fruit}")

print("Modified list of fruits:", fruit_list)
### END CODE
```

Exercise 6 (4 points)
Write a Python program and get tuples of car names. Then intruduce a second tuple that is the first tuples * n. Print your second tuple.

```python
### YOUR CODE GOES BELOW
cars = input("Enter car names separated by commas: ")

car_tuple = tuple(cars.split(","))

n = int(input("Enter a number to repeat the tuple: "))

second_tuple = car_tuple * n

print("Second tuple:", second_tuple)
### END CODE
```
