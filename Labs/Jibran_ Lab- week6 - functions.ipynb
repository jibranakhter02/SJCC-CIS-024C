
1. Write a Python function to find the maximum of three numbers.

```python
def max_of_three(a, b, c):
    return max(a, b, c)

print(max_of_three(5, 10, 3))
```

2. Write a Python program to reverse a string.
Sample String : "1234abcd"

```python
def reverse_string(s):
    return s[::-1]

sample_string = "1234abcd"
print(reverse_string(sample_string))
```

3. Write a Python function that accepts a string and counts the number of upper and lower case letters.

Sample String : 'The quick Brow Fox'

Expected Output :

No. of Upper case characters : 3

No. of Lower case Characters : 12

```python
def count_case(s):
    upper_case = 0
    lower_case = 0

    for char in s:
        if char.isupper():
            upper_case += 1
        elif char.islower():
            lower_case += 1

    print("No. of Upper case characters:", upper_case)
    print("No. of Lower case characters:", lower_case)

sample_string = 'The quick Brow Fox'
count_case(sample_string)
```

4. Write a Python function that checks whether a passed string is a palindrome or not.
Note: A palindrome is a word, phrase, or sequence that reads the same backward as forward, e.g., madam or nurses run.

```python
def is_palindrome(s):
    normalized_string = ''.join(s.split()).lower()
    return normalized_string == normalized_string[::-1]

# Test the function
print(is_palindrome("madam"))
print(is_palindrome("nurses run"))
print(is_palindrome("hello"))
```

5. Write a Python function that takes a list and returns a new list with distinct elements from the first list.
Sample List : [1,2,3,3,3,3,4,5]

Unique List : [1, 2, 3, 4, 5]

```python
def unique_elements(lst):
    return list(set(lst))

sample_list = [1, 2, 3, 3, 3, 3, 4, 5]
unique_list = unique_elements(sample_list)
print("Unique List:", unique_list)
```
