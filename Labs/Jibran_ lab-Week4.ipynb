#### Exercise 3 (20 points)

```python
### YOUR CODE GOES BELOW
age_1 = int(input("Enter the age of user_1: "))
age_2 = int(input("Enter the age of user_2: "))
age_3 = int(input("Enter the age of user_3: "))
age_4 = int(input("Enter the age of user_4: "))

average_age = (age_1 + age_2 + age_3 + age_4) / 4
print(f"The average age is: {average_age:.2f}")

if age_1 == age_2 or age_1 == age_3 or age_1 == age_4:
    print(f"user_1 has the same age as another user.")
if age_2 == age_3 or age_2 == age_4:
    print(f"user_2 has the same age as another user.")
if age_3 == age_4:
    print(f"user_3 has the same age as user_4.")

ages = [("user_1", age_1), ("user_2", age_2), ("user_3", age_3), ("user_4", age_4)]
ages.sort(key=lambda x: x[1])

print(f"{ages[3][0]} is the oldest with age {ages[3][1]}.")
print(f"{ages[0][0]} is the youngest with age {ages[0][1]}.")
### END CODE
```

#### Exercise 4 (20 points)

Write a Python program that ask users (4 users: name_1, ...) to write their first names.

* Check the first letter of each name and

if it is a vowel print "the first letter of name_x is a vowel"

if it is not a vowel print "the first letter of name_x is a constant"

```python
### YOUR CODE GOES BELOW
name_1 = input("Enter the first name of name_1: ")
name_2 = input("Enter the first name of name_2: ")
name_3 = input("Enter the first name of name_3: ")
name_4 = input("Enter the first name of name_4: ")

def check_first_letter(name):
    if name[0].lower() in 'aeiou':
        return f"The first letter of {name} is a vowel."
    else:
        return f"The first letter of {name} is a consonant."

print(check_first_letter(name_1))
print(check_first_letter(name_2))
print(check_first_letter(name_3))
print(check_first_letter(name_4))
### END CODE
```

#### Exercise 5 (20 points)


Write a Python program that ask user to choose (5 colos: color_1, ...).

Check how many colors are in dark group and how many are in light group

dark group = black, blue, brown, green, purple, red, burgendy

light color = yellow, orange, pink, creame, white, gray

Then print "color_x and ... is in light group "
also print "color_y and ... is in dark group "


```python
### YOUR CODE GOES BELOW
dark_group = {"black", "blue", "brown", "green", "purple", "red", "burgundy"}
light_group = {"yellow", "orange", "pink", "cream", "white", "gray"}

user_input = input("Enter 5 colors separated by commas: ")
colors = [color.strip().lower() for color in user_input.split(",")]

dark_colors = list(filter(lambda color: color in dark_group, colors))
light_colors = list(filter(lambda color: color in light_group, colors))

if light_colors:
    print(f"{', '.join(light_colors)} is in the light group.")
else:
    print("No colors are in the light group.")

if dark_colors:
    print(f"{', '.join(dark_colors)} is in the dark group.")
else:
    print("No colors are in the dark group.")
### END CODE
```
