#### Exercise 1 - Accessing values in a dictionary object

Initialize a dictionary object with the below key, value pairs

 Key  | Value
 -----|------
'Harry'|'Teacher'
'Sally'|'Scientist'
'Joe'|'Farmer'
'Mathew'|'Detective'

Use a while loop to ask the user to enter a name, and display the corresponding value. If the name is not found, display "Could not find user"

At the end of each iteration, ask the user if they want to continue. If the user wants to continue, repeat the question, otherwise, terminate the while loop and display the message "Program ended"

**Example**
```
Enter a name: Harry
Occupation is Teacher

Do you want to continue(y/n): y

Enter a name: Joe
Occupation is Farmer

Enter a name: Mark
Could not find user

Do you want to continue(y/n): n
Program ended
```

````python
### YOUR CODE GOES BELOW
occupations = {
    'Harry': 'Teacher',
    'Sally': 'Scientist',
    'Joe': 'Farmer',
    'Mathew': 'Detective'
}

while True:
    name = input("Enter a name: ")

    if name in occupations:
        print(f"Occupation is {occupations[name]}")
    else:
        print("Could not find user")

    continue_program = input("Do you want to continue (y/n): ").lower()

    if continue_program == 'n':
        print("Program ended")
        break
### END CODE
````

#### Exercise 2

Write a Python script to check if a given key already exists in a dictionary

**Hint**: Use the **get** method to determine if a key exists

````python
### YOUR CODE GOES BELOW
sample_dict = {
    'Harry': 'Teacher',
    'Sally': 'Scientist',
    'Joe': 'Farmer',
    'Mathew': 'Detective'
}

key_to_check = input("Enter a key to check: ")

if sample_dict.get(key_to_check) is not None:
    print(f"The key '{key_to_check}' exists with value '{sample_dict[key_to_check]}'.")
else:
    print(f"The key '{key_to_check}' does not exist.")
### END CODE
````

#### Exercise 3

Write a Python script to create a dictionary that contains a sequence of numbers from 1 to n and their squares. Ask the user to enter n. Display the resulting dictionary.

**Example**
If user enters 5, then your dictionary will look like

{1:1,2:4,3:9:4:16,5:25}

**Hint** You will need to assign new values to the dictionary to create the dictionary object

````python
### YOUR CODE GOES BELOW
n = int(input("Enter a number: "))

squares_dict = {}

for i in range(1, n+1):
    squares_dict[i] = i ** 2

print(squares_dict)
### END CODE
````

#### Exercise 4

Refer to Exercise 3 for this problem.

Write a python program to sum all the values in the above dictionary (n:n^2). Display the resulting sum


````python
### YOUR CODE GOES BELOW
n = int(input("Enter a number: "))

squares_dict = {}

for i in range(1, n+1):
    squares_dict[i] = i ** 2

sum_of_squares = sum(squares_dict.values())

print(f"The dictionary is: {squares_dict}")
print(f"The sum of all values is: {sum_of_squares}")
### END CODE
````

#### Exercise 5

Write a Python program to sort the below dictionary by key in descending order.

 {2:"Mary",1:"Sally",5:"Harry",4:"Joe"}

**Hint:** Refer to the class example

````python
### YOUR CODE GOES BELOW
sample_dict = {2: "Mary", 1: "Sally", 5: "Harry", 4: "Joe"}

sorted_dict = dict(sorted(sample_dict.items(), key=lambda item: item[0], reverse=True))

print("Dictionary sorted by key in descending order:")
print(sorted_dict)
### END CODE
````

#### Exercise 6
Write a python program to combine the below two dictionaries and display the resulting dictionary

* Dictionary 1:  {2:"Mary",1:"Sally",5:"Harry",4:"Joe"}
* Dictionary 2:  {6:"John",9:"Nancy",12:"Peter",8:"Alice"}


````python
### YOUR CODE GOES BELOW
dict1 = {2: "Mary", 1: "Sally", 5: "Harry", 4: "Joe"}
dict2 = {6: "John", 9: "Nancy", 12: "Peter", 8: "Alice"}

combined_dict = {**dict1, **dict2}

print("Combined dictionary:")
print(combined_dict)
### END CODE
````
