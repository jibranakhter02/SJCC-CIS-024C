#### Exercise 2 (10 points)

Write a Python program to use for loop to find the factorial of a given number.

The factorial (symbol: !) means multiplying all numbers from the chosen number down to 1.

For example, a factorial of 5! is 5 × 4 × 3 × 2 × 1 = 120

Expected output:

120

```python
### YOUR CODE GOES BELOW
number = int(input())
f=1
for i in range(1,number+1):
  f*=i

print("the factorial is: ", f)



### END CODE
```

#### Exercise 3 (10 points)

Display Fibonacci series up to 12 terms.

Fibonacci Sequence is a series of numbers in which the next number is found by adding up the two numbers before it. The first two numbers are 0 and 1.

For example, 0, 1, 1, 2, 3, 5, 8, 13, 21. The next number in this series is 13 + 21 = 34.


```python
### YOUR CODE GOES BELOW
n_terms = 12

n1, n2 = 0, 1
count = 0

if n_terms <= 0:
    print("Please enter a positive integer.")
elif n_terms == 1:
    print("Fibonacci sequence up to", n_terms, ":")
    print(n1)
else:
    print("Fibonacci sequence up to", n_terms, ":")
    while count < n_terms:
        print(n1, end=" ")
        nth = n1 + n2
        n1 = n2
        n2 = nth
        count += 1
### END CODE
```

#### Exercise 4 (10 points)

Write a Python program to display only those numbers from a list that satisfy the following conditions

  1-The number must be divisible by five
  
  2-If the number is greater than 150, then skip it and move to the following number

  3-If the number is greater than 500, then stop the loop



**Given:**

numbers = [12, 75, 150, 180, 145, 525, 50]

**Expected output:**

75
150
145





```python
### YOUR CODE GOES BELOW
numbers = [12, 75, 150, 180, 145, 525, 50]

for num in numbers:
    if num > 500:
        break
    if num > 150:
        continue
    if num % 5 == 0:
        print(num, end=" ")
### END CODE
```
